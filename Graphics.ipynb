```python
from subprocess import Popen
import time
grad_time = []
metro_time = []
metro_an_time = []

my_range = range(1,100) #[2, 5, 10, 35, 50, 70, 90, 100]
my_second_range = range(15)
for i in my_range:
    test_time = time.time()
    try:
        process = Popen('/home/axcel/Документы/VertexCover/grad.out ' + str(i) ,
                    shell=True)
        process.wait()
    except Exception:
        process.terminate()
    grad_time.append(time.time() - test_time)
    print(i)
        
```

```python
for i in my_range:
    test_time = time.time()
    try:
        process = Popen('/home/axcel/Документы/VertexCover/metro.out ' + str(i) ,
                    shell=True)
        process.wait()
    except Exception:
        process.terminate()
    metro_time.append(time.time() - test_time)   
    print(i)      
```

```python
for i in my_range:
    test_time = time.time()
    try:
        process = Popen('/home/axcel/Документы/VertexCover/metro_an.out ' + str(i) ,
                    shell=True)
        process.wait()
    except Exception:
        process.terminate()
    metro_an_time.append(time.time() - test_time)
    print(i)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import pylab
from pylab import rcParams

plt.figure(figsize=(9,5))
plt.plot(my_range, grad_time, label='gradient')
plt.plot(my_range, metro_an_time, label='metropolis with annealing')
plt.plot(my_range, metro_time, label='metropolis')
plt.xlabel('Vertexes')
plt.ylabel('Time')

pylab.legend(('gradient','metropolis with annealing','metropolis'), loc=2)
plt.show()
```
